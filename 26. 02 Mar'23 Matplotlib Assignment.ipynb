## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np<br>
np.random.seed(3)<br>
x = 3 + np.random.normal(0, 2, 50)<br>
y = 3 + np.random.normal(0, 2, len(x))<br>
Note: Also add title, xlabel, and ylabel to the plot.

`Answer`

A scatter plot is a type of plot that displays values for typically two variables as points on a two-dimensional plane. It shows the relationship between the variables and helps to identify any patterns or trends in the data.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:<br>
import numpy as np<br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])<br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])<br>

`Answer`

The `subplot()` function in matplotlib is used to create multiple plots within a single figure. It allows you to arrange multiple subplots in a grid-like structure, making it convenient to compare and analyze different datasets or variables.

```python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2)
# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("Line 1")

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title("Line 2")

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title("Line 3")

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title("Line 4")

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np<br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"])<br>
profit = np.array([3000, 8000, 1000, 10000])<br>

`Answer`

A bar plot, also known as a bar chart or bar graph, is a visual representation of categorical data using rectangular bars. It is used to compare and display the values of different categories or groups. The length or height of each bar corresponds to the magnitude of the data it represents.

Bar plots are commonly used to:

1. Compare data between different categories or groups.
2. Show the distribution or frequency of categorical data.
3. Display trends or changes over time.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
# Horizontal bar plot
plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)<br>

`Answer`

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays summary statistics such as the minimum, first quartile, median, third quartile, and maximum values, as well as any potential outliers.

Box plots are used for the following purposes:

1. Visualizing the distribution of data: Box plots provide a concise summary of the distribution, showing the spread, skewness, and presence of outliers.

2. Comparing distributions: Box plots allow for easy comparison of multiple datasets or groups, making it useful for identifying differences in medians, ranges, and variabilities.

3. Detecting outliers: Box plots highlight any values that are considered outliers, helping to identify potential anomalies or extreme observations.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Values")
plt.show()
```
