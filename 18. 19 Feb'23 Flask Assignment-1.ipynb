## Q1. What is Flask Framework? What are the advantages of Flask Framework?

`Answer`

Flask is a micro web framework written in Python. It is designed to be simple and lightweight, with a modular approach to web development. Flask allows developers to easily build web applications using Python by providing a minimalistic set of tools and libraries.

The advantages of Flask Framework are:

1. Lightweight and flexible: Flask is a lightweight framework, which means that it is easy to learn and use. It is also highly flexible, allowing developers to customize their applications to meet their specific needs.

2. Modular design: Flask has a modular design, which means that it allows developers to choose the components they want to use and customize their applications accordingly. This makes it easy to add or remove features as needed.

3. Easy to set up and use: Flask is easy to set up and use, making it a popular choice for developers who want to quickly build web applications.

4. Large community and ecosystem: Flask has a large and active community of developers, which means that it is well-supported and has a wide range of plugins and extensions available.

5. Built-in development server: Flask comes with a built-in development server, which makes it easy to test and debug applications during the development process.

6. Integrates well with other tools: Flask integrates well with other tools and libraries, including SQLAlchemy, Jinja2, and WTForms, making it a versatile choice for web development.

Overall, Flask is a powerful and flexible web framework that provides developers with a simple and elegant solution for building web applications using Python.

## Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return '<h1>Hello World!!</h1>'


if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

## Q3. What is App routing in Flask? Why do we use app routes?

`Answer`

In Flask, an app route is a way to bind a URL to a Python function or view. In other words, when a user visits a specific URL in their browser, Flask will execute the associated Python function to generate and return the response to the user.

We use app routes in Flask for a number of reasons:

1. Routing URLs: By using app routes, we can map specific URLs to specific functions or views in our Flask application. This allows users to access different pages or functionality of our application by visiting different URLs.

2. Separating concerns: By breaking up our application into different routes and functions, we can better separate the concerns of our application. For example, we may have one route to handle user authentication, another route to handle user profile information, and so on.

3. Easier maintenance: With app routes, it's easier to maintain our application and make changes as needed. We can add, modify, or remove routes and functions without affecting the rest of our application.

4. Better organization: Using app routes also helps us organize our code and make it more modular. We can break our application into different components that handle specific tasks, which makes our code easier to understand and maintain.

Overall, app routes are an essential part of building web applications with Flask, as they allow us to handle different URLs and serve different content to users based on their requests.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: 
* Company Name: ABC Corporation 
* Location: India 
* Contact Detail: 999-999-9999

## Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_info():
    company_name = 'ABC Corporation'
    location = 'India'
    contact = '999-999-9999'
    
    return f'Company Name: {company_name}<br>Location: {location}<br>Contact Detail: {contact}'

if __name__ == '__main__':
    app.run(host="0.0.0.0")    
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

`Answer`

In Flask, the url_for() function is used for URL building. It is a helper function that generates a URL to a specific function dynamically. This means that if you change the name of a function or the URL pattern associated with it, the url_for() function will automatically generate a new URL based on the updated information.

Here's an example Python code to demonstrate the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/user/<username>')
def profile(username):
    return f'Hi there, {username}!'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('about')) 
    print(url_for('profile', username='manish')) 
```
